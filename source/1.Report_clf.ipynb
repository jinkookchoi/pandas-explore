```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from IPython.display import display, HTML, Markdown, Latex, Image
import pandas_explore as exp
```

```python
# Titanic data
df_titanic = pd.read_csv('./data/titanic.csv').drop(columns=['Name', 'Cabin', 'Ticket'])
df_titanic = df_titanic.astype({'Survived':'object', 'Pclass':'object'})
df_titanic_train, df_titanic_test = train_test_split(df_titanic, test_size=0.2)
df_titanic_test = df_titanic_test.drop(columns=['Survived'])
```

# A. Classification 모델의 단일 데이터 프레임 리포트

```python
exp.report_dataset(
    {'set_name': 'Titanic', 'dfs': [{'df_name': 'Train', 'df': df_titanic_train}]}, 
    model_type='clf', target_col= 'Survived'
)
```

# B. Regression 모델의 2개의 데이터 프레임 리포트

```python
exp.report_dataset(
    {'set_name': 'Titanic', 'dfs': [{'df_name': 'Train', 'df': df_titanic_train}, {'df_name': 'Test', 'df': df_titanic_test}]}, 
    model_type='clf', target_col= 'Survived'
)
```
