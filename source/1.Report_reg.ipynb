```python
import numpy as np
import pandas as pd
import pandas_explore as exp
from sklearn.model_selection import train_test_split
from IPython.display import display, HTML, Markdown, Latex, Image
```

```python
df_diamonds = pd.read_csv('../data/diamonds.csv')
df_dia_train, df_dia_test = train_test_split(df_diamonds, test_size=0.2)
df_dia_test = df_dia_test.drop(columns=['price'])
```

# A. Regression 모델의 단일 데이터 프레임 리포트

```python
exp.report_dataset(
    {'set_name': 'Diamonds', 'dfs': [{'df_name': 'Train', 'df': df_dia_train}]}, 
    model_type='reg', target_col= 'price'
)
```

# B. Regression 모델의 2개의 데이터 프레임 리포트

```python
exp.report_dataset(
    {'set_name': 'Diamonds', 'dfs': [{'df_name': 'Train', 'df': df_dia_train}, {'df_name': 'Test', 'df': df_dia_test}]}, 
    model_type='reg', target_col= 'price'
)
```
